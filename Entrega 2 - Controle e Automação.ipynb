## O Método do Lugar das Raízes (Capítulo 7)

O Método do Lugar das Raízes (Root Locus Method), introduzido por Walter R. Evans em 1948, representa um marco fundamental na teoria de controle clássico. Ele fornece ao engenheiro uma representação visual qualitativa e quantitativa do desempenho do sistema de controle em malha fechada à medida que um parâmetro do sistema — tipicamente o ganho de laço, $K$ — é variado.1 A premissa central é que a resposta transitória de um sistema linear invariante no tempo (LIT) é ditada predominantemente pela localização de seus polos no plano $s$. Portanto, compreender a trajetória desses polos é compreender a evolução do comportamento do sistema.

### Fundamentação Matemática e Conceitual


A equação característica de um sistema de controle com realimentação negativa é a base algébrica para o método. Considere um sistema com função de transferência de malha aberta $L(s) = G(s)H(s)$. A função de transferência de malha fechada $T(s)$ é dada por:

$$T(s) = \frac{G(s)}{1 + G(s)H(s)}$$

Os polos de malha fechada são as raízes da equação característica:

$$1 + K G(s)H(s) = 0$$

Para que um ponto complexo $s$ pertença ao lugar das raízes, ele deve satisfazer a equação vetorial complexa derivada acima. Esta equação pode ser decomposta em duas condições independentes que devem ser satisfeitas simultaneamente: a Condição de Ângulo e a Condição de Módulo.

#### A Condição de Ângulo

A condição de ângulo é o critério determinante para a existência do lugar geométrico. Visto que $K$ é um escalar real positivo ($K > 0$), a fase de $K$ é $0^\circ$. Assim, rearranjando a equação característica para $K G(s)H(s) = -1$, observa-se que a fase do termo $G(s)H(s)$ deve ser um múltiplo ímpar de $180^\circ$ para compensar o sinal negativo (que é um vetor unitário a $180^\circ$ no plano complexo).

Matematicamente, para um ponto $s$ pertencer ao lugar das raízes, a soma dos ângulos dos vetores que partem dos zeros de malha aberta até $s$, subtraída da soma dos ângulos dos vetores que partem dos polos de malha aberta até $s$, deve resultar em um múltiplo ímpar de $\pi$ radianos:

$$ \angle G(s)H(s) = \sum_{i=1}^{m} \angle(s + z_i) - \sum_{j=1}^{n} \angle(s + p_j) = (2k + 1)180^\circ, \quad k = 0, \pm 1, \pm 2, \dots $$

Esta condição é geométrica e independente do valor de $K$. Ela define as "trilhas" no plano $s$ onde as raízes podem existir.

### Procedimento Analítico de Construção do Lugar das Raízes

A construção do lugar das raízes segue um algoritmo sistemático derivado das condições de ângulo e módulo. Este procedimento permite ao engenheiro esboçar a topologia das raízes sem resolver numericamente a equação característica para infinitos valores de $K$.

#### Análise Aprofundada das Assíntotas

A configuração das assíntotas é frequentemente o primeiro indicador da viabilidade de um controle proporcional simples. Se $n - m \ge 3$, pelo menos uma assíntota invariavelmente entrará no semi-plano direito (SPD) para ganhos elevados, indicando que o sistema se tornará instável. Isso impõe um limite superior prático ao ganho $K$ utilizável, restringindo o desempenho (velocidade de resposta e erro estacionário) que pode ser obtido sem compensação adicional.

#### Pontos de Ramificação (Breakaway Points)

Os pontos de ramificação no eixo real são de extrema importância física. Eles representam a condição de amortecimento crítico ($\zeta = 1$). Ao ajustar $K$ para operar exatamente neste ponto, obtém-se a resposta mais rápida possível sem oscilação (sobressinal). O cálculo exato desses pontos envolve a diferenciação da função de malha aberta, o que pode resultar em polinômios de alta ordem, exigindo métodos numéricos para solução precisa.

### Projeto de Parâmetros e Especificações de Desempenho

O "Lugar das Raízes" não é apenas uma ferramenta de análise de estabilidade, mas um método robusto de síntese. O projeto de parâmetros via Root Locus envolve a seleção de um ponto $s_d$ (polo dominante desejado) sobre as trajetórias existentes que satisfaça as especificações de desempenho no domínio do tempo. As principais métricas de projeto correlacionam-se geometricamente no plano $s$ da seguinte forma:

1. **Sobressinal Percentual ($PO$):** Diretamente relacionado à razão de amortecimento $\zeta$. No plano $s$, linhas de $\zeta$ constante são raios que partem da origem. O ângulo $\theta$ desse raio com o eixo real negativo é $\theta = \cos^{-1}(\zeta)$. Portanto, especificar um $PO$ máximo equivale a definir uma região proibida abaixo de uma certa linha radial.
2. **Tempo de Assentamento ($T_s$):** Aproximadamente $T_s \approx 4/(\zeta\omega_n)$ (critério de 2%). O produto $\zeta\omega_n$ é a parte real do polo ($\sigma$). Assim, especificar $T_s$ máximo define uma linha vertical $\sigma = -4/T_s$ no plano $s$. Os polos devem estar à esquerda desta linha.
3. **Tempo de Pico ($T_p$) e Tempo de Subida ($T_r$):** Relacionados à frequência natural amortecida $\omega_d = \omega_n\sqrt{1-\zeta^2}$, que é a coordenada imaginária do polo.

O procedimento de projeto consiste em desenhar a "região de projeto" definida por essas fronteiras e verificar se o lugar das raízes atravessa essa região. Se atravessar, calcula-se o ganho $K$ no ponto de interseção usando a condição de módulo. Caso contrário, o sistema requer compensação dinâmica (adição de polos e zeros) para remodelar o lugar das raízes.

### Controladores PID no Contexto do Lugar das Raízes

O capítulo dedica atenção especial à implementação de controladores PID (Proporcional-Integral-Derivativo), analisando como cada termo altera a topologia do lugar das raízes. A função de transferência de um controlador PID ideal é:

$$G_c(s) = K_P + \frac{K_I}{s} + K_D s = \frac{K_D s^2 + K_P s + K_I}{s}$$

Esta estrutura introduz um polo na origem (devido à ação integral) e dois zeros (que podem ser reais ou complexos conjugados) no sistema.

- **Efeito da Ação Derivativa (PD):** Um controlador PD ($G_c(s) = K_D s + K_P$) adiciona um zero simples ao sistema. No lugar das raízes, a adição de um zero tende a "puxar" os ramos do lugar das raízes para a esquerda (em direção ao semi-plano estável). Isso ocorre porque os ramos tendem a terminar nos zeros. Fisicamente, isso resulta em um aumento do amortecimento e da estabilidade, permitindo ganhos mais altos e respostas mais rápidas.
- **Efeito da Ação Integral (PI):** Um controlador PI ($G_c(s) = K_P + K_I/s$) adiciona um polo na origem e um zero finito. O polo na origem aumenta o "Tipo" do sistema, eliminando o erro em regime permanente para entradas degrau. No entanto, no lugar das raízes, a adição de um polo na origem tende a empurrar os ramos para a direita, reduzindo a estabilidade relativa. O zero do PI é usado para mitigar esse efeito desestabilizador, devendo ser colocado próximo ao polo na origem para minimizar a distorção no lugar das raízes das raízes dominantes (configuração de dipolo).
- **Sintonia PID:** O método de Ziegler-Nichols e a sintonia manual são discutidos como formas de encontrar os parâmetros iniciais, mas o refinamento via lugar das raízes permite uma visualização exata de como os zeros do controlador afetam o overshoot e o tempo de assentamento.1

### Sensibilidade e Variação Paramétrica

A análise de sensibilidade no contexto do lugar das raízes (Seção 7.5) quantifica a robustez do sistema. A sensibilidade da raiz $r_i$ em relação a um parâmetro $\alpha$ (que pode ser o ganho $K$ ou a posição de um polo da planta $p$) é definida como:

$$S_\alpha^{r_i} = \frac{\partial r_i}{\partial \alpha / \alpha}$$

Sistemas onde os polos de malha fechada se movem drasticamente com pequenas variações nos parâmetros da planta são considerados pouco robustos. O lugar das raízes permite visualizar essas sensibilidades: regiões onde os ramos estão muito próximos do eixo imaginário ou onde a densidade de raízes varia rapidamente com $K$ indicam baixa robustez.

### Lugar das Raízes com Ganho Negativo

Embora menos comum na prática de sistemas convencionais estáveis, o capítulo aborda o Lugar das Raízes de Ganho Negativo ($K < 0$), relevante para sistemas com realimentação positiva ou plantas com ganho negativo intrínseco. A modificação crucial reside na condição de ângulo:

$$\angle G(s)H(s) = k 360^\circ \quad (\text{ou } 0^\circ)$$

Isso inverte as regras de existência no eixo real (o lugar existe à esquerda de um número par de polos e zeros) e altera os ângulos das assíntotas. A análise complementar é vital para garantir a estabilidade global em sistemas complexos ou multivariáveis onde o sinal do ganho efetivo pode variar.

## O Projeto de Sistemas de Controle com Realimentação (Capítulo 10)

Enquanto o Capítulo 7 foca na análise e ajuste de ganho, o Capítulo 10 trata da compensação, ou seja, a alteração intencional da estrutura do sistema através da adição de dinâmicas (polos e zeros) para atingir especificações de desempenho que o simples ajuste de ganho não consegue satisfazer. O capítulo concentra-se em compensadores em cascata: Avanço (Lead), Atraso (Lag) e Avanço-Atraso (Lead-Lag), utilizando tanto a abordagem de Bode quanto a de Lugar das Raízes.

### Compensação em Cascata: Conceitos Fundamentais

A compensação em cascata envolve a inserção de um compensador $G_c(s)$ em série com a planta $G(s)$. A função de transferência de malha aberta torna-se $L(s) = G_c(s)G(s)$. O objetivo é remodelar $L(s)$ para que a malha fechada atenda a requisitos de:

1. Precisão em regime permanente (erro estático).

2. Resposta transitória (tempo de subida, sobressinal).

3. Estabilidade relativa (margem de fase e ganho).

4. Rejeição de distúrbios.

### Projeto de Compensador de Avanço de Fase (Phase-Lead)

O compensador de avanço é essencialmente um filtro passa-alta que introduz fase positiva no sistema. Sua função de transferência é:

$$ G_c(s) = \frac{s + z}{s + p} = \frac{1}{\alpha} \frac{s + \frac{1}{\tau}}{s + \frac{1}{\alpha\tau}}, \quad \text{onde } |z| < |p| \text{ (ou } \alpha < 1) $$

O zero está localizado mais próximo da origem do que o polo.

#### Abordagem via Diagrama de Bode

No domínio da frequência, o objetivo do compensador de avanço é aumentar a Margem de Fase (PM) do sistema. A adição de fase positiva na frequência de cruzamento de ganho ($\omega_{cg}$) aumenta o amortecimento do sistema em malha fechada.

**Procedimento de Projeto (Bode):**

1. Determine o ganho $K$ necessário para satisfazer o erro em regime permanente.
2. Avalie a Margem de Fase do sistema não compensado com este ganho.
3. Determine o avanço de fase necessário ($\phi_m$) para atingir a PM desejada, adicionando uma margem de segurança (5° a 12°) para compensar o deslocamento da frequência de cruzamento.
4. Calcule o fator de atenuação $\alpha$ usando a relação:$$ \sin \phi_m = \frac{1 - \alpha}{1 + \alpha} \implies \alpha = \frac{1 - \sin \phi_m}{1 + \sin \phi_m} $$
5. Determine a frequência onde o máximo avanço de fase deve ocorrer ($\omega_m$). Esta frequência deve se tornar a nova frequência de cruzamento de ganho. Como o compensador altera a magnitude, a nova frequência de cruzamento é encontrada onde a magnitude do sistema não compensado é igual a $-10 \log(1/\alpha)$ dB
6. Calcule o zero ($z = 1/\tau$) e o polo ($p = 1/(\alpha\tau)$) usando $\omega_m = \frac{1}{\tau\sqrt{\alpha}}$.

**Implicações:** O compensador de avanço aumenta a largura de banda do sistema, resultando em uma resposta mais rápida (menor tempo de subida), mas também aumenta a suscetibilidade a ruídos de alta frequência devido ao ganho elevado nessa região.

#### Abordagem via Lugar das Raízes

No domínio do tempo, o compensador de avanço é usado para remodelar o lugar das raízes de modo que ele passe por um par de polos dominantes desejados ($s_d$) que não pertenciam ao lugar original.

**Procedimento de Projeto (LGR):**

1. Determine a localização desejada dos polos de malha fechada ($s_d$) baseada em $\zeta$ e $\omega_n$.
2. Calcule a deficiência angular ($\phi$) no ponto $s_d$. A soma dos ângulos dos polos e zeros da planta original em $s_d$ não será um múltiplo ímpar de $180^\circ$. A diferença é o ângulo que o compensador deve fornecer:$$\angle G_c(s_d) = 180^\circ - \angle G(s_d)H(s_d)$$
3. Posicione o polo e o zero do compensador. Existem infinitas soluções, mas uma prática comum é colocar o zero do compensador logo abaixo do polo desejado (cancelamento) ou usar o método da bissetriz para equilibrar a contribuição angular.
4. O zero atrai o lugar das raízes para a esquerda, aumentando a estabilidade e a velocidade de resposta.

### Projeto de Compensador de Atraso de Fase (Phase-Lag)

O compensador de atraso é um filtro passa-baixa que atenua o ganho em altas frequências. Sua função é:

$$G_c(s) = \frac{s + z}{s + p}, \quad \text{onde } |p| < |z|$$

O polo está mais próximo da origem do que o zero.

#### Abordagem via Diagrama de Bode

O objetivo primário aqui não é adicionar fase, mas sim permitir um aumento no ganho DC (para reduzir erro) sem comprometer a estabilidade.



**Procedimento de Projeto (Bode):**

1. Ajuste o ganho $K$ para satisfazer o erro estático (isso geralmente desestabiliza o sistema ou reduz drasticamente a PM).
2. Encontre a frequência onde a fase do sistema não compensado é igual à margem de fase desejada (mais uma margem de segurança). Essa será a nova frequência de cruzamento $\omega'_{cg}$.
3. O compensador de atraso deve prover uma atenuação na magnitude para trazer a curva de ganho para 0 dB em $\omega'_{cg}$. A quantidade de atenuação necessária determina a relação entre polo e zero.
4. Posicione o zero e o polo uma década (ou mais) abaixo de $\omega'_{cg}$ para evitar que o atraso de fase inerente do compensador degrade a margem de fase na frequência de cruzamento.

**Implicações:** O compensador de atraso reduz a largura de banda do sistema, resultando em uma resposta transitória mais lenta, mas melhora drasticamente o erro em regime permanente e rejeita ruídos de alta frequência.

#### Abordagem via Lugar das Raízes

No LGR, o compensador de atraso é projetado como um dipolo (polo e zero muito próximos) próximo à origem.



- A proximidade entre polo e zero garante que a contribuição angular e de magnitude no lugar das raízes (região transiente) seja quase nula, preservando a forma do lugar das raízes original.
- No entanto, a relação $z/p$ pode ser grande (ex: 10 ou 100). Como o ganho estático do compensador é $K_c(z/p)$, isso amplifica o ganho de erro do sistema sem deslocar significativamente os polos dominantes.
- Regra: $|p| < |z| \ll |\text{polos dominantes}|$.

### Compensação Avanço-Atraso (Lead-Lag) e Integração

Quando é necessário melhorar tanto a resposta transitória quanto o erro estacionário, utiliza-se o compensador Avanço-Atraso.

- A parte de Avanço aumenta a largura de banda e a margem de fase.
- A parte de Atraso aumenta o ganho em baixa frequência.
- Matematicamente, são dois compensadores em série: $G_c(s) = \left(\frac{s+z_1}{s+p_1}\right)\left(\frac{s+z_2}{s+p_2}\right)$.

O capítulo também discute Compensadores de Integração (como o termo I no PID ou um atraso puro), que aumentam o tipo do sistema, garantindo erro zero para degraus ou rampas, mas introduzindo um atraso de fase de $90^\circ$ que deve ser gerido.

### Pré-Filtros e Resposta Deadbeat

Uma técnica avançada discutida é o uso de um Pré-Filtro ($G_p(s)$) fora do laço de feedback. Isso permite cancelar zeros de malha fechada que causam sobressinal excessivo sem alterar as propriedades de rejeição de perturbação do laço. O projeto para Resposta Deadbeat é introduzido como um ideal de controle: atingir o alvo no menor tempo possível com erro zero e sem oscilação, um conceito que faz a ponte para o controle digital.

## Sistemas de Controle Digital (Capítulo 13)

O Capítulo 13 marca a transição do controle analógico contínuo para o controle discreto, refletindo a onipresença de microprocessadores, DSPs e computadores na automação moderna. Neste paradigma, os sinais não são contínuos no tempo, mas amostrados em intervalos discretos $T$.1

### Sistemas de Dados Amostrados e o Processo de Amostragem

Um sistema de dados amostrados híbrido contém sinais contínuos e discretos. O elemento central é o Amostrador (Sampler). Matematicamente, a amostragem ideal é modelada como a modulação do sinal contínuo $r(t)$ por um trem de impulsos (delta de Dirac) periódico:

$$r^*(t) = \sum_{k=0}^{\infty} r(kT) \delta(t - kT)$$

Onde $T$ é o período de amostragem. No domínio da frequência, a amostragem cria réplicas espectrais do sinal original a cada $\omega_s = 2\pi/T$. Isso introduz o fenômeno de aliasing se o Teorema de Nyquist ($\omega_s > 2\omega_{max}$) não for respeitado.

Para reconstruir o sinal contínuo a partir das amostras (necessário para controlar uma planta analógica), utiliza-se um Segurador de Ordem Zero (Zero-Order Hold - ZOH). O ZOH mantém o valor da amostra constante durante o intervalo $T$. Sua função de transferência é fundamental para a modelagem de sistemas digitais:

$$G_{ho}(s) = \frac{1 - e^{-sT}}{s}$$

O ZOH introduz um atraso de fase aproximado de $T/2$, que pode desestabilizar o sistema se $T$ for muito grande.

### A Transformada Z

Como a Transformada de Laplace de um sinal amostrado resulta em uma soma infinita de exponenciais $e^{-skT}$, ela é intratável para análise algébrica direta. A Transformada Z é introduzida como uma mudança de variável fundamental:

$$z = e^{sT}$$

Esta definição mapeia o plano $s$ complexo no plano $z$ complexo:

- **Eixo Imaginário ($j\omega$):** Mapeia-se no Círculo Unitário ($|z|=1$).
- **Semi-plano Esquerdo (Estável):** Mapeia-se no Interior do Círculo Unitário ($|z| < 1$).
- **Semi-plano Direito (Instável):** Mapeia-se no Exterior do Círculo Unitário ($|z| > 1$).

A definição formal da Transformada Z de uma sequência $f(kT)$ é:

$$F(z) = \mathcal{Z}\{f(t)\} = \sum_{k=0}^{\infty} f(kT) z^{-k}$$

Esta transformada converte equações de diferenças (que descrevem algoritmos computacionais) em equações algébricas em $z$, permitindo o uso de funções de transferência.13

### Função de Transferência de Pulso

A função de transferência de pulso, $G(z)$, relaciona a entrada e a saída nos instantes de amostragem. Para um sistema contínuo $G(s)$ precedido por um ZOH, a função de transferência discreta equivalente é calculada como:

$$
G(z) = \mathcal{Z}\left\{ \frac{1 - e^{-sT}}{s} \, G(s) \right\}
= (1 - z^{-1}) \, \mathcal{Z}\left\{ \frac{G(s)}{s} \right\}
$$


Esta fórmula é crucial para discretizar modelos de plantas contínuas para simulação e projeto de controle digital.

### Análise de Estabilidade no Plano Z

A estabilidade de um sistema digital é determinada pelas raízes da equação característica em $z$: $1 + G(z) = 0$ (para um sistema simples). Para estabilidade assintótica, todas as raízes devem estar estritamente dentro do círculo unitário ($|z| < 1$).

#### Critério de Estabilidade de Jury

Análogo ao critério de Routh-Hurwitz para o plano $s$, o Teste de Jury permite verificar se existem raízes fora do círculo unitário sem calculá-las explicitamente. O teste envolve a construção de uma tabela baseada nos coeficientes do polinômio característico $D(z) = a_n z^n + a_{n-1} z^{n-1} + \dots + a_0$ e a verificação de condições de desigualdade. Se qualquer condição falhar, o sistema é instável.14

### Desempenho e Projeto no Domínio Z

Os conceitos de erro e resposta transitória são mapeados para o plano $z$.

**Erro em Regime Estacionário:** O Teorema do Valor Final para transformadas Z é:

$$\lim_{k \to \infty} e(kT) = \lim_{z \to 1} (z - 1) E(z)$$

Isso permite calcular erros de posição, velocidade e aceleração para sistemas digitais.



**Controladores PID Digitais:** A implementação de controladores PID em computadores requer a aproximação numérica de integrais e derivadas. O capítulo discute métodos como a Diferença para Trás (Backward Difference), onde $s \approx \frac{z-1}{zT}$, e a aproximação trapezoidal (Tustin). Isso transforma a função de transferência contínua do PID em uma equação de diferenças implementável em código:

$$u[k] = u[k-1] + K_1 e[k] + K_2 e[k-1] + K_3 e[k-2]$$
